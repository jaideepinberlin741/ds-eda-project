```python
import warnings

warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

house_details_df = pd.read_csv("data/kids-density_central_locality_new.csv")

house_sales_df = pd.read_csv("data/eda_king_county_house_sales_cleaned.csv")


house_details_df.head()

waterfront_houses = house_details_df[house_details_df['waterfront'] == 1.0]


# Get the count of houses having waterfront
waterfront_count = house_details_df[house_details_df['waterfront'] == 1.0].shape[0]
print(f"Number of houses with waterfront: {waterfront_count}")

waterfront_houses.head()
```

```python
import plotly.express as px

fig = px.scatter_mapbox(
    house_details_df[house_details_df['bedrooms'] >= 2.0],
    lat="lat",
    lon="long",
    center=dict(lat=house_details_df["lat"].mean(), lon=house_details_df["long"].mean()),
    zoom=9,
    mapbox_style="open-street-map",  # kein Token nÃ¶tig
)

fig.show()
```

```python
# import data frames to work on 
import warnings

warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_sales_merged = pd.merge(house_details_df, house_sales_df, on='house_id', how='inner')

house_sales_waterfront = df_sales_merged[(df_sales_merged['waterfront'] == 1) & (df_sales_merged['bedrooms'] > 1)]

fig_larry = px.scatter_mapbox(house_sales_waterfront[house_sales_waterfront['price'] < 800000], 
                              lat='lat', 
                              lon='long', 
                              color='larry_index', 
                              size='price', 
                              title='Larry Index for Waterfront Houses',
                              color_continuous_scale='Viridis',
                              mapbox_style='open-street-map',
                              hover_data=['house_id', 'price', 'bedrooms', 'bathrooms', 'sqft_living', 'larry_index', 'centrality_seattle', 'population_density', 'kid_density'],
                              zoom=9)
fig_larry.show()
```
